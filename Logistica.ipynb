```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import *
```

```python
x=arange(1,100,0.1)
a=0.07
c0=0.2
```

```python
plot(c0*exp(a*x))
```

```python
for i in arange(0.1,1,0.1):
    plot(i*exp(a*x))
```

```python
def f(x,t):
    return a*x
```

```python
x0=0.1
T=arange(1,200,0.1)
for i in arange(0.1,1,0.1):
    sim=odeint(f,i,T)
    plot(sim)
show()
```

```python
k=20
def g(x,t):
    return a*x*(1-(x/k))
```

```python
x0=0.1
T=arange(1,200,0.1)
for i in arange(0.5,40,4):
    sim=odeint(g,i,T)
    plot(sim)
show()
```

```python
def C(x0):
    return x0/(1-x0/k)
def G(x0,t):
    return (C(x0)*exp(a*t))/(1+(C(x0)/k)*exp(a*t))
```

```python
for i in arange(0.5,40,0.8):
    plot(G(i,T))
```

### El Oscilador armónico

Consideramos la ecuación

$$
\frac{d^2x}{dt^2}+\omega_0^2 x=2\zeta \omega_0 \frac{dx}{dt}
$$

equivalentemente tenemos el sistema

$$
\frac{dx}{dt}=y\\
\frac{dy}{dt}=-\omega_0^2x+2\zeta \omega_0y
$$

$\omega_0$ es la frecuencia del oscilador

Para distintos valores de $\zeta$ tenemos los casos siguientes:

* Sin Amortiguamiento para $\zeta=0$
* Sub-amortiguado para $0<\zeta<1$
* Críticamente amortiguado para $\zeta=1$
* Sobre-amortiguado para $\zeta>1$

```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import *
```

```python
def dy(y, t, zeta, w0):
    x, p = y[0], y[1]
    dx = p
    dp = -2 * zeta * w0 * p - w0**2 * x
    return [dx, dp]


```

```python
# Condicion inicial
y0 = [1.0, 0.0]

# Vector de tiempos
t = linspace(0, 10, 1000)
w0 = 2*pi

```

```python
###SOLUCION
y1 = odeint(dy, y0, t, args=(0.0, w0)) # sin amortiguamiento
y2 = odeint(dy, y0, t, args=(0.2, w0)) # sub-amortiguado
y3 = odeint(dy, y0, t, args=(1.0, w0)) # criticamente amortiguado
y4 = odeint(dy, y0, t, args=(5.0, w0)) # sobre-amortiguado
```

```python
plot(y2.T[0])
```

```python

fig=figure(figsize=(20,10))
plot(t, y1[:,0], 'y', label="sin amortiguamiento")
plot(t, y2[:,0], 'r', label="sub-amortiguado")
plot(t, y3[:,0], 'b', label=r"cr$\rm{\'i}$ticamente amortiguado")
plot(t, y4[:,0], 'g', label="sobre-amortiguado")
title(r"Oscilador arm$\rm{\'o}$nico")
legend()
show()
```

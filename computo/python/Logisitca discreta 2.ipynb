```python
from numpy import*
from matplotlib.pyplot import *
```

```python
X0=0.1
X=[X0]
```

```python
def f(r,x):
    return r*x*(1-x)
```

```python
def f2(r,x):
    return f(r,f(r,x))
def f3(r,x):
    return f(r,f(r,f(r,x)))
```

```python
T=arange(0,1.01,0.01)
r=4
plot(T,f(r,T),T,f2(r,T),T,f3(r,T),T,T)
```

```python
r=3.8
plot(T,f2(r,T),label='r=%1f'%r)
plot(T,T,label='y=x')
xc=0.73
for i in range(1000):
    plot([xc,xc],[xc,f2(r,xc)], color='k', linestyle='--')
    plot([f2(r,xc),xc],[f2(r,xc),f2(r,xc)], color='k', linestyle='--')
    xc=f2(r,xc)
legend()
```

```python
r=3.6
plot(T,f3(r,T),label='r=%1f'%r)
plot(T,T,label='y=x')
xc=0.5
for i in range(1000):
    plot([xc,xc],[xc,f3(r,xc)], color='k', linestyle='--')
    plot([f3(r,xc),xc],[f3(r,xc),f3(r,xc)], color='k', linestyle='--')
    xc=f3(r,xc)
legend()
```

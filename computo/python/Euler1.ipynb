```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t,y):
    return r*y
```

```python
h=0.5
T=np.arange(0,10,h)
r=0.2
```

```python
sol=[0.1]
for t in T:
    y=sol[-1]+h*f(t,sol[-1])
    sol.append(y)
```

```python
plt.plot(sol)
```

```python
S=0.1*np.exp(r*T)
```

```python
plt.plot(S,label='exacta')
plt.plot(sol,label='euler 0.5')
plt.legend()
```

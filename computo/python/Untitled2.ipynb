# Ecuaciones diferenciales con Sage

```python
from sage.calculus.desolvers import desolve_odeint
```

**Ecuaci√≥nes de Volterra**

Consideramos el sistema no-lineal

$$
\dot{x}=ax-bxy\\
\dot{y}=dxy-cy
$$

```python
x,y=var('x,y')
f=[x*(1-y),-y*(1-x)]
sol=desolve_odeint(f,[0.5,2],srange(0,10,0.1),[x,y])
p=line(zip(sol[:,0],sol[:,1]))
```

```python
p.show()
```

**Ecuaciones de Lorenz**

El sistema

$$
\dot{x}=\sigma(y-x)\\
\dot{y}=x(\rho-z)-y\\
\dot{z}=xy-\beta z
$$

```python
x,y,z=var('x,y,z')
sigma=10
rho=28
beta=8/3
lorenz=[sigma*(y-x),x*(rho-z)-y,x*y-beta*z]
times=srange(0,50.05,0.01)
ics=[0,1,1]
sol=desolve_odeint(lorenz,ics,times,[x,y,z],rtol=1e-13,atol=1e-14)
```

```python
p=line(zip(sol[:,0],sol[:,1],sol[:,2]))
q=line(zip(sol[:,0],sol[:,1]))
r=line(zip(sol[:,1],sol[:,2]))
s=line(zip(sol[:,0],sol[:,2]))           
```

```python
p.show(viewer='tachyon')
q.show()
r.show()
s.show()
```

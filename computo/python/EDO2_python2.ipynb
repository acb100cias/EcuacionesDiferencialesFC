```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import odeint
```

```python
def f(x,t):
    return r*x*(1-x/k)
```

```python
r=0.2
k=20.

x0=5.

T=arange(0,100,0.01)
```

```python
T
```

```python
X=odeint(f,x0,T).T[0]
```

```python
plot(T,X)
```

```python
for ci in [1,2,5,10,15,20,25,30,35]:
    X=odeint(f,ci,T)
    plot(T,X,label='%d'%ci)
    legend()
```

```python
h=0.05
def g(x,t):
    return r*x*(1-x/k)-h*x
```

```python
x=5.
X=odeint(g,x0,T).T[0]
```

```python
plot(T,X)
```

```python
for ci in [1,2,5,10,15,20,25,30,35]:
    X=odeint(g,ci,T)
    plot(T,X,label='%d'%ci)
    legend()
```

```python
h=1.5
def g2(x,t):
    return r*x*(1-x/k)-h*sin(pi*t)
```

```python
x0=25
X=odeint(g2,x0,T).T[0]
plot(T,X)
```

```python
a=1.
b=0.1
c=1.5
d=0.75
def dX(X,t):
    return array([a*X[0]-b*X[0]*X[1],
                  -c*X[1]+d*X[0]*X[1]])

T=arange(0,100,0.001)
X0=array([10,5])
```

```python
X=odeint(dX,X0,T).T
```

```python
X[1]
```

```python
plot(T,X[0],label='Presa')
plot(T,X[1],label='Depredador')
legend()
```

```python
plot(X[0],X[1])
```

```python
beta=0.1
gamma=0.02
def dX(X,t):
    return array([-beta*X[0]*X[1],
                  beta*X[0]*X[1]-gamma*X[2],
                  gamma*X[2]])

T=arange(0,200,1)
X0=array([4700-10,10,0])
```

```python
X=odeint(dX,X0,T).T
```

```python
plot(T,X[0])
plot(T,X[1])
plot(T,X[2])
```

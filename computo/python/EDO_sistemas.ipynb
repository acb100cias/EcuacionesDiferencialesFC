```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import odeint
```

```python
def F(y,t,N,beta,gamma):
    S,I,R=y
    dS=-beta*S*I/N
    dI=beta*S*I/N-gamma*I
    dR=gamma*I
    return dS,dI,dR
```

```python
N=120000000
I0=100
R0=0
S0=N-I0-R0

beta=0.6
gamma=0.3

t=linspace(0,365,365)
y0=S0,I0,R0

Sol=odeint(F,y0,t,args=(N,beta,gamma))
```

```python
S,I,R=Sol.T
```

```python
plot(t,S,'g-',label='Suceptibles')
plot(t,I,'r-',label='Infecciosos')
plot(t,R,'k-',label='Removidos')
legend()
```

```python
B=arange(0.1,1.,0.2)
C=arange(0.02,0.1,0.04)
figure(figsize=(20,10))
for b in B:
    for c in C:
        Sol=odeint(F,y0,t,args=(N,b,c))
        S,I,R=Sol.T
        plot(S,I,label=r'$\beta=$%f,$\gamma=$%f'%(b,c))
xlabel('S')
ylabel('I')
legend(loc='upper right')
title('Plano fase de SIR')
```

```python
mort=0.05
gamma=0.01
Sol=odeint(F,y0,t,args=(N,beta,gamma))
S,I,R=Sol.T
Re=R*(1-mort)
M=R-Re
```

```python
plot(t,S,'g-',label='suceptibles')
plot(t,I,'r-',label='infecciosos')
plot(t,Re,'b-',label='recuperados')
plot(t,M,'k-',label='muertos')
legend()
```

```python
def L(X,t,s,r,b):
    x,y,z =X
    dx=s*(y-x)
    dy=r*x-y-x*z
    dz = x*y -b*z
    return dx,dy,dz
```

```python
x0,y0,z0=(0,1,1.05)
```

```python
t=arange(0,10000,0.01)
l0=[x0,y0,z0]
s=10
r=28
b=2.667
```

```python
S2=odeint(L,l0,t,args=(s,r,b))
```

```python
x,y,z=S2.T
```

```python
fig=figure()
ax=fig.gca(projection='3d')
ax.plot(x,y,z)
```

# Metodo de Runge-Kutta

$x(t_0)=x_0$

$k_1=f(t_i,x_i)$

$k_2=f(t_i+\frac{1}{2}h,x_i+\frac{1}{2}k_1h)$

$k_3=f(t_i+\frac{1}{2}h,x_i+\frac{1}{2}k_2h)$

$k_4=f(t_i+h,x_i+k_3h)$

$x_{i+1}=x_i+\frac{1}{6}h(k_1+2k_2+2k_3+k_4)$




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t,x):
    return r*x
```

```python
r=0.2
x0=1
h=1.5
T=np.arange(0,100,h)
```

```python
solRK=[]
solRK.append(x0)
for t in T[1:]:
    k1=f(t,solRK[-1])
    k2=f(t+h/2,solRK[-1]+(h/2*k1))
    k3=f(t+h/2,solRK[-1]+(h/2*k2))
    k4=f(t+h,solRK[-1]+h*k3)
    x=solRK[-1]+(h*1/6)*(k1+2*k2+2*k3+k4)
    solRK.append(x)
```

```python
plt.plot(T,solRK,'o',markersize=6,label='RK')
S=x0*np.exp(r*T)
plt.plot(T,S,'o',markersize=3,label='exacta')
plt.legend()
```

```python
from numpy import  *
from matplotlib.pyplot import *
from scipy.integrate import*
```

```python
a=1.9
def f(x,t):
    return a*x*(1-x)
x0=0.2
T=arange(0,30,0.001)
sim=odeint(f,x0,T)
plot(T,sim)
```

```python
for i in arange(0.1,2,0.2):
    x0=i
    T=arange(0,30,0.001)
    sim=odeint(f,x0,T)
    plot(T,sim)
```

```python
a=1.9
h=0.5
def f(x,t):
    return a*x*(1-x)-h*sin(2*pi*t)
x0=1.2
T=arange(0,30,0.001)
sim=odeint(f,x0,T)
plot(T,sim)
```

```python
a=1.9

def f(x,t):
    return a*x*(1-(x/h))-(x**2)/(1+x**2)
x0=1.2
T=arange(0,30,0.001)
sim=odeint(f,x0,T)
def g(x,h):
    return (1-(x/h))-(x**2)/(1+x**2)
figure()
plot(T,sim)
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(arange(-10,10,0.1),arange(-10,10,0.1),array([list(map(g,arange(-10,10,0.1),arange(-10,10,0.1))),(list(map(g,arange(-10,10,0.1),arange(-10,10,0.1))]))
```

```python
# parametros
a = 1.
b = 0.1
c = 1.5
d = 0.75
def dX_dt(X, t=0):
    return array([ a*X[0] -   b*X[0]*X[1] ,
                  -c*X[1] + d*b*X[0]*X[1] ])
t = arange(0, 15,  0.001) #tiempo              
X0 = array([10, 5]) #ci                    
X = odeint(dX_dt, X0, t)
pres, dep = X.T
print(X)
f1 = figure()
plot(t, pres, 'r-', label='Presa')
plot(t, dep  , 'b-', label='Depredador')
legend(loc='best')
xlabel('time')
ylabel('population')
title('Solucion Presa-depredador')
#f1.savefig('presa-depredador')
```

```python
# parametros
a = 1.
b = 0.1
c = 1.5
d = 0.75
e=0.2
f=0.5
def dX_dt(X, t=0):
    return array([ a*X[0]*(1-e*X[0]) -   b*X[0]*X[1] ,
                  -c*X[1]*(1-f*X[1]) + d*b*X[0]*X[1] ])
t = arange(0, 15,  0.001) #tiempo              
X0 = array([0.02, 0.4]) #ci                    
X = odeint(dX_dt, X0, t)
pres, dep = X.T
print(X)
f1 = figure()
plot(t, pres, 'r-', label='Presa')
plot(t, dep  , 'b-', label='Depredador')
legend(loc='best')
xlabel('time')
ylabel('population')
title('Solucion Presa-depredador')
```

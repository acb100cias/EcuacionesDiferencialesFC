```python
from numpy import *
from matplotlib.pyplot import*
from scipy.integrate import*
```

```python
def s(t,c):
    return (200*t)/10. +c*(200-t)**(26)
```

```python
x=arange(0,200,0.1)
c=0.
plot(s(x,c))
```

```python
def f(s,t):
    return (5./2)-(26*s)/(200-t)
```

```python
x0=0
T=arange(0,200,0.1)
sim=odeint(f,x0,T)
plot(T,sim)
show()
```

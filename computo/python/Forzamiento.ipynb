```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import *

```

```python
def f(Y, t):
    y1, y2 = Y
    return [y2, -k*y1-b*y2-g(t,w)]
def g(w,t):
    return cos(w*t)


```

```python
k=2.
b=0
w=0.05
```

```python
for y10 in [0, 0.5, 1, 1.5, 2, 2.5]:
    tspan = linspace(0, 30, 20000)
    y0 = [y10, 0]
    ys = odeint(f, y0, tspan)
    plot(ys[:,0], ys[:,1], 'b-') # path
    plot([ys[0,0]], [ys[0,1]], 'o',label='%f'%y10) # start
    plot([ys[-1,0]], [ys[-1,1]], 's') # end
    

xlim([-4, 4])
xlabel('$y_1$')
ylabel('$y_2$')
title('Trayectorias en el espacio-fase')
legend()
```

```python
def A(q,p,w):
    return 1./(sqrt((q-w**2)**2+p**2*w**2))
```

```python
q=2.
w=arange(0,3,0.1)
for p in [0.01,0.1,0.5,1.,2.0]:
    plot(w,A(q,p,w),label='%f'%p)
    xlabel('w')
    ylabel('A')
    legend()
    title('Amplitud de la oscilacion')
```

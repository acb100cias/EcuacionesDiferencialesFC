# Modelos de combate

## PRIMEROS MODELOS DE COMBATE 

Hace poco más de cien años, el teniente de la Marina de los EE. UU. (Más tarde contralmirante) J.V. Chase desarrolló una ecuación diferencial para modelar los enfrentamientos entre dos flotas homogéneas. Aproximadamente una década después, el ingeniero británico Frederick W. Lanchester llegó de forma independiente a una ecuación casi idéntica, con ejemplos que cubren también el combate aéreo y terrestre. Un ruso contemporáneo de ambos, M. Osipov, también llegó a conclusiones similares en un artículo publicado en la revista zarista "Military Digest" (Voennyi Sbornik) en 1915.

De los tres, la historia ha sido más amable con Lanchester: el trabajo de Chase no fue desclasificado hasta 1972, demasiado tarde para que él disfrutara del reconocimiento público que merecía; y Osipov, aunque a veces se menciona junto a su homólogo británico, sigue siendo un extraño para nosotros.

En aplicaciones prácticas, los modelos de combate simples descritos por Lanchester y sus contemporáneos se superan en su mayoría en la actualidad, en parte debido a la naturaleza más compleja y dinámica del campo de batalla moderno, pero principalmente debido al poder de cálculo que trajo la era de la computadora, que hace mucho posibles simulaciones más sofisticadas.

Y, sin embargo, las leyes de Lanchester, tal como las conocemos hoy, nunca parecen perder relevancia. Muchos modelos de combate introducidos en las últimas décadas se derivan directamente de ellos (COMAN, Bonder-Farrell), o los utilizan a nivel local para describir enfrentamientos más pequeños dentro de un marco más amplio.

# LEY LINEAL DE LANCHESTER

Lanchester comenzó postulando una condición simple pero razonable de combate antiguo (formaciones cercanas de hombres que usan espadas o lanzas) en lo que llamó la "Ley Lineal": un combatiente armado con un arma cuerpo a cuerpo solo puede atacar un objetivo a la vez, dentro del brazo alcanzar. Lo que esto significa es que las bajas que una fuerza podría sufrir o causar en un instante dado tienen un límite superior práctico, proporcional al número de sus combatientes que están en contacto físico directo con el enemigo. A lo sumo, este límite superior sería el número de unidades activas en la fuerza menor, pero el terreno o el frente de formación podrían imponer más limitaciones. Podemos hacer las siguientes observaciones de estos dos compromisos:

La tasa de desgaste de ambos ejércitos es constante durante toda la batalla (excepto al final, cuando quedaron menos soldados de los que encajarían en el frente original)
Un ejército más grande no confiere un mayor poder de matar, solo un mayor poder de permanencia, ya que el ejército más grande puede absorber más bajas antes de ser eliminado.
Un ejército que es superado en número dos a uno puede compensarlo siendo exactamente el doble de letal. En otras palabras, la letalidad y los números tienen el mismo peso a la hora de decidir el resultado de una batalla.
El frente no altera realmente el resultado de la batalla, solo su duración.
Según Lanchester, se podrían aplicar exactamente las mismas reglas a un escenario moderno de fuego sin tiempo, en el que ambos lados disparaban a un enemigo que no podían ver, y simplemente barrían un área objetivo constante: en tal caso, el lado más grande habría mayor potencia de fuego, pero esto se compensaría con una menor densidad de objetivos enemigos; el bando superado en número tendría menos armas, pero sus disparos encontrarían su objetivo con más frecuencia. El resultado sería una tasa de desgaste constante para ambos ejércitos, exactamente como en los ejemplos de combate cuerpo a cuerpo anteriores.

# Ley cuadrada de Lanchester

En el combate moderno con armas de largo alcance y fuego dirigido, continuó Lanchester, la restricción establecida por la Ley Lineal se levanta: ahora, cualquier combatiente puede atacar a cualquier objetivo dentro del alcance y, a su vez, recibir fuego de múltiples enemigos. Por lo tanto, la tasa de desgaste ya no está limitada por el frente: todos los elementos pueden participar en todas las etapas de la batalla.

En estas condiciones, si un ejército tenía 20 elementos de combate, y se esperaba que cada uno eliminara una unidad enemiga en promedio cada dos "pasos de tiempo" de un encuentro dado (lo que sea que realmente entendamos por un "paso de tiempo", ya sea un minuto o al día, depende del alcance de nuestra simulación), entonces se esperaría que el ejército eliminara 10 enemigos en promedio por paso de tiempo. En un caso general, un ejército con varios elementos A, todos los cuales participan en el encuentro en un paso de tiempo, siempre causará un número de bajas a un ejército B de A multiplicado por α, donde α es el número esperado de bajas. por elemento por unidad de tiempo en promedio. Esto se traduce en las siguientes ecuaciones:

$$
\frac{dB}{dt} = -\alpha A\\
\frac{dA}{dt} = -\beta B
$$

Donde A y B son el número de soldados en cada ejército, y α y ß su número promedio esperado de muertes por soldado por unidad de tiempo. Hoy, nos referimos a α y ß como coeficientes de tasa de deserción de Lanchester.

Apliquemos estas ecuaciones a nuestro ejemplo anterior de un ejército azul que cuenta con 42 soldados y un ejército rojo con una fuerza de 30. Como antes, hacemos que ambos ejércitos sean exactamente igual de letales; se espera que ambos, en este ejemplo, neutralicen a uno. enemigo cada cinco pasos de tiempo (1/5 = 0,2 muertes por paso de tiempo). Sustituyendo estos valores en las ecuaciones anteriores, obtenemos:

$$
\frac{dA}{dt} = -0.2 \times 42\\
\frac{dB}{dt} = -0.2 \times 30
$$

En nuestro ejemplo de la Ley Lineal, con los mismos números, el ejército Azul ganó con solo 12 unidades supervivientes. Esta vez, aplicando la Ley del Cuadrado, las cosas están considerablemente más torcidas, ya que 29 elementos de Azul sobreviven ilesos.

Intuitivamente podemos apreciar cómo, en las condiciones establecidas por la Ley del Cuadrado de Lanchester, la superioridad numérica tiene un efecto mayor en el resultado de un encuentro que la letalidad de cada unidad individual: cada baja infligida no solo reduce la cantidad de enemigos que pueden contraatacar. , pero también aumenta la concentración de fuego en cada objetivo en el siguiente pulso. Un acuerdo tácito con la máxima militar de Nathan Bedford Forrest "llegar primero con la mayoría de los hombres".

Analíticamente, podemos dividir las ecuaciones para las tasas de accidentes, obteniendo:

$$
\frac{dA}{dB} = \frac{\beta B}{\alpha A}
$$

Que podemos reescribir como:

$$
\alpha A \times dA = \beta B \times dB
$$

para relacionar las pérdidas instantáneas de ambos ejércitos, y luego integrar para encontrar que:
$$
\alpha (A_0^2 - A^2) = \beta(B_0^2 - B^2)
$$

Lo que muestra que la fuerza de combate de un ejército es proporcional al cuadrado de su tamaño, de ahí la "Ley del cuadrado".

# Validación de la ley cuadrada de Lanchester:

Como modelo de combate, la ley cuadrada de Lanchester tiene algunas limitaciones considerables, entre otras:

* Asume una inteligencia perfecta, ya que ambos lados conocen la ubicación exacta del enemigo en todo momento.
* Asume un control de fuego perfecto, ya que ambos lados pueden concentrar el fuego de la manera más eficiente.
* Asume que las condiciones del campo de batalla son constantes, ya que los coeficientes de eficiencia α y ß no se ven afectados por cambios en el terreno, maniobras, moral, suministros, etc.
* Es un modelo determinista, ya que la misma entrada siempre dará el mismo resultado.
* Es un modelo homogéneo, en el que se supone que todos los elementos de una unidad de combate tienen las mismas características de combate.

Muchos autores han propuesto revisiones y expansiones del trabajo de Lanchester para corregir algunas de estas deficiencias, o para tener en cuenta una variedad de factores como refuerzos y retiradas, pérdidas operativas, mando y control, guerra de guerrillas e insurgencia, los efectos supresores de las armas, cerca de apoyo aéreo, etc. Una lista detallada de estos desarrollos hasta finales de la década de 1970 se puede encontrar en Lanchester-Type Models of Warfare de James G. Taylor, vol. Yo, 1980.

Dicho esto, debemos considerar la Ley del Cuadrado no como un modelo de combate, sino como un modelo de desgaste. En otras palabras, no podemos esperar que describa una vista detallada del campo de batalla en evolución, sino más bien que describa el proceso matemático mediante el cual se tratan y mantienen las bajas, una vez que se conocen todos los demás factores.

Y en ese sentido, puede hacerlo sorprendentemente bien. Aunque existen otros estudios que ilustran claramente el asunto, J.H. **Implementación de la ley lineal de Lanchester**

```python
import numpy

import matplotlib.pyplot as plot

from math import ceil

# The length of the time step will not alter the end result.
# Use only to determine the resolution of the graph.

timeStart = 0.0
timeEnd = 10.0
timeStep = 0.01

steps = int((timeEnd - timeStart) / timeStep)

# Initialise numpy arrays covering each step of the simulation.

blue = numpy.zeros(steps)
red = numpy.zeros(steps)
time = numpy.zeros(steps)

# To remove the frontage constraint, change the frontage variable to
# the smaller remaining force, both in its declaration and in the loop.

blue[0] = 42
red[0] = 30
frontage = 5

blueLethality = 1
redLethality = 1

time[0] = timeStart

for i in range(steps -1):
    frontage = min(frontage, ceil(red[i]), ceil(blue[i]))
    blue[i+1] = max(0, blue[i] - timeStep * (frontage * redLethality))
    red[i+1] = max(0, red[i] - timeStep * (frontage * blueLethality))
    time[i+1] = time[i] + timeStep
    
# Remaining forces at the end of the simulation, for plot label purposes.
    
blueRemaining = int(blue[len(blue)-1])
redRemaining = int(red[len(red)-1])

# Plot code.
    
plot.figure()
plot.step(time, blue, '-b', where = 'post', label = 'Blue army')
plot.step(time, red, '-r', where = 'post', label = 'Red army')
plot.ylabel('Strength')
plot.xlabel('Time')
plot.legend()
plot.annotate(blueRemaining,
              xy=(timeEnd, blue[len(blue)-1]),
              xytext=(-15,10),
              textcoords='offset points')
plot.annotate(redRemaining,
              xy=(timeEnd, red[len(red)-1]),
              xytext=(-15,10),
              textcoords='offset points')
plot.show()
```

**Implementación de la ley cuadrada de Lanchester**

```python
import numpy

import matplotlib.pyplot as plot

# The length of the time step will not alter the end result.
# Use only to determine the resolution of the graph.

timeStart = 0.0
timeEnd = 5.0
timeStep = 0.01

steps = int((timeEnd - timeStart) / timeStep)

# Initialise numpy arrays covering each step of the simulation.

blue = numpy.zeros(steps)
red = numpy.zeros(steps)
time = numpy.zeros(steps)

# Iwo Jima sample values: blue (US) = 54000; red (Japanese) = 21500;
# blueLethality = 0.0106; redLethality = 0.0544

blue[0] = 42
red[0] = 30

blueLethality = 0.2
redLethality = 0.2

time[0] = timeStart

for i in range(steps -1):
    blue[i+1] = max(0, blue[i] - timeStep * (red[i] * redLethality))
    red[i+1] = max(0, red[i] - timeStep * (blue[i] * blueLethality))
    time[i+1] = time[i] + timeStep
    
# Remaining forces at the end of the simulation, for plot label purposes.
    
blueRemaining = int(blue[len(blue)-1])
redRemaining = int(red[len(red)-1])

# Plot code.
    
plot.figure()
plot.step(time, blue, '-b', where = 'post', label = 'Blue army')
plot.step(time, red, '-r', where = 'post', label = 'Red army')
plot.ylabel('Strength')
plot.xlabel('Time')
plot.legend()
plot.annotate(blueRemaining,
              xy=(timeEnd, blue[len(blue)-1]),
              xytext=(-15,10),
              textcoords='offset points')
plot.annotate(redRemaining,
              xy=(timeEnd, red[len(red)-1]),
              xytext=(-15,10),
              textcoords='offset points')
plot.show()
```

# Método de Euler.

Aproximaremos numéricamente las soluciones de una ecuación diferencial usando el esquema iterativo 
$$
\frac{\mathrm{d}y}{\mathrm{d}x}=f(x,y)
$$


$$
y_0=y(x_0) 
$$
$$
y_1=y_0+hf(x_0,y_0)
$$
$$
y_k=y_{k-1}+hf(x_{k_1},y_{k-1})
$$

```python
from numpy import *
from matplotlib.pyplot import *
```

```python
def f(x,y):
    return 0.1*y

def euler(x,y,f,h):
    return y+h*f(x,y)
```

```python
x0=0.1
y0=0.1

Y=[y0]
X=arange(x0,10,0.01)

for i in range(len(X)-1):
    y=euler(X[i],Y[-1],f,0.1)
    Y.append(y)
F=[y0]
for i in range(len(X)-1):
    F.append(y0*exp(X[i]))
    
        
    
    
```

```python
E=[]
for i in range(len(X)):
    E.append((abs(F[i]+Y[i])/abs(F[i])))
    
```

```python
plot(X,E)
```

```python
plot(X,Y,label='aprox')
plot(X,F,label='$y_0e^{ax}$')
legend()
```

```python
k=20.
a=0.1
def g(x,y):
    return a*y*(1-(y/k))
def C(x0):
    return x0/(1-(x0/k))
def G(x0,t):
    return (C(x0)*exp(a*t))/(1+(C(x0)/k)*exp(a*t))
```

```python
x0=0.1
y0=2

Y=[y0]
X=arange(x0,100,0.1)

for i in range(len(X)-1):
    y=euler(x0,Y[-1],g,0.1)
    Y.append(y)
    
F=[x0]
for i in range(len(X)-1):
    F.append(G(y0,X[i]))
```

```python
plot(X,Y,'.',label='Euler')
plot(X,F,'.',label='Exacta')
legend()
```

```python
def Q(x,y):
    return x+y
```

```python
x0=0.1
y0=1

Y=[y0]
X=arange(x0,20,0.1)

for i in range(len(X)-1):
    y=euler(X[i],Y[-1],Q,0.1)
    Y.append(y)
F=[x0]
for i in range(len(X)-1):
    F.append(2*exp(X[i]) - X[i] - 1)
    
```

```python
plot(X,Y,label='aprox')
plot(X,F,label='$2e^{x}-x-1$')
legend()
```
